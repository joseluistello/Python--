#### Mmmm... our numbers looks ugly. Let's fix it. 

```python
jejeje = dfcashflow
```

```python
jejeje.astype('category')
```

```python
format_dict = {col_name: '${:,.2f}' for col_name in jejeje.select_dtypes(include=['int64']).columns}

format_dict['Individual % of total'] = '{:.1%}'

jejeje.head().style.format(format_dict)
```

```python
sns.set_style('dark')
ax = sns.lineplot(
            x='netIncome', 
            y='date',
            data=jejeje,
            color = 'c'
                );
ax.set(xlabel='Year', ylabel='Total Inventory in Billion USD')
plt.show()
```
