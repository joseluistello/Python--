```python
date = []
symbol = []
revenue = []
costOfRevenue = []
netIncome = []

     
for finance in data:
    date.append(finance["date"])
    symbol.append(finance["symbol"])
    revenue.append(finance["revenue"])
    costOfRevenue.append(finance["costOfRevenue"])
    netIncome.append(finance["netIncome"])

      
finance_dict = {
    "Date" : date,
    "Ticket": symbol,
    "Revenue" : revenue,
    "CostOfRevenue" : costOfRevenue,
    "NetIncome" : netIncome
}
```
